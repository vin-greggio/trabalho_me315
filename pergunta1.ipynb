```python
## Pega a precipitação de cada estado da região NORTE em cada ano:

resultados_norte = []
for i in range(2000, 2025):
    query =f'''
    SELECT SUM("PRECIPITAÇÃO TOTAL, HORÁRIO (mm)") AS precipitacao_total,
           SUM("PRECIPITAÇÃO TOTAL, HORÁRIO (mm)")/COUNT(DISTINCT estacao) AS precipitacao_por_estacao,
           estado, regiao
    FROM dados_{i} LEFT JOIN estacoes ON estacao = codigo_estacao
    WHERE "PRECIPITAÇÃO TOTAL, HORÁRIO (mm)" IS NOT NULL
    AND regiao = "N"
    GROUP BY regiao, estado
    '''
    resultados_norte.append((f'dados_{i}',pl.read_database(query, conn)))

print(resultados_norte)
```

```python
import pandas as pd
```

```python
### TABELA DE PRECIPITAÇÃO TOTAL ###

#calcular a média anual de precipitação -> encontrar o maior e o menor por região
def calcular_maior_menor_precipitacao(resultados):
    #dicionário: armazena a soma total e a contagem de anos por estado
    precipitacao_media_por_estado = {}

    #precipitação total de cada estado
    for ano, df in resultados:
        for linha in df.iter_rows(named=True):
            estado = linha["estado"]
            precipitacao_total = linha["precipitacao_total"]
            if estado not in precipitacao_media_por_estado:
                precipitacao_media_por_estado[estado] = {"precipitacao_total": 0, "anos": 0}
            precipitacao_media_por_estado[estado]["precipitacao_total"] += precipitacao_total
            precipitacao_media_por_estado[estado]["anos"] += 1

    #média anual para cada estado
    medias_por_estado = {estado: dados["precipitacao_total"] / dados["anos"]
                         for estado, dados in precipitacao_media_por_estado.items()}

    #estado com maior e menor precipitação média
    estado_maior = max(medias_por_estado, key=medias_por_estado.get)
    estado_menor = min(medias_por_estado, key=medias_por_estado.get)

    return estado_maior, medias_por_estado[estado_maior], estado_menor, medias_por_estado[estado_menor]

#resultados para cada região
tabela_final = []

regioes_dados = {
    "Norte": resultados_norte,
    "Nordeste": resultados_nordeste,
    "Centro-Oeste": resultados_centro_oeste,
    "Sudeste": resultados_sudeste,
    "Sul": resultados_sul
}

for regiao, resultados in regioes_dados.items():
    estado_maior, maior_precipitacao, estado_menor, menor_precipitacao = calcular_maior_menor_precipitacao(resultados)
    tabela_final.append([regiao, estado_maior, maior_precipitacao, estado_menor, menor_precipitacao])

#tabela final
tabela_final_df = pd.DataFrame(tabela_final, columns=["Região", "Estado Maior Precipitação Média", "Maior Precipitação Média (mm)", "Estado Menor Precipitação Média", "Menor Precipitação Média (mm)"])
print(tabela_final_df)
```

```python
### TABELA DE PRECIPITAÇÃO POR ESTAÇÃO ###

#calcular a média anual de precipitação por estação -> encontrar o maior e o menor por região
def calcular_maior_menor_precipitacao(resultados):
    #dicionário: armazena a soma total e a contagem de anos por estado
    precipitacao_media_por_estado = {}

    #precipitação total de cada estado por estação
    for ano, df in resultados:
        for linha in df.iter_rows(named=True):
            estado = linha["estado"]
            precipitacao_por_estacao = linha["precipitacao_por_estacao"]
            if estado not in precipitacao_media_por_estado:
                precipitacao_media_por_estado[estado] = {"precipitacao_por_estacao": 0, "anos": 0}
            precipitacao_media_por_estado[estado]["precipitacao_por_estacao"] += precipitacao_por_estacao
            precipitacao_media_por_estado[estado]["anos"] += 1

    #média anual para cada estado
    medias_por_estado = {estado: dados["precipitacao_por_estacao"] / dados["anos"]
                         for estado, dados in precipitacao_media_por_estado.items()}

    #estado com maior e menor precipitação média
    estado_maior = max(medias_por_estado, key=medias_por_estado.get)
    estado_menor = min(medias_por_estado, key=medias_por_estado.get)

    return estado_maior, medias_por_estado[estado_maior], estado_menor, medias_por_estado[estado_menor]

#resultados para cada região
tabela_final = []

regioes_dados = {
    "Norte": resultados_norte,
    "Nordeste": resultados_nordeste,
    "Centro-Oeste": resultados_centro_oeste,
    "Sudeste": resultados_sudeste,
    "Sul": resultados_sul
}

for regiao, resultados in regioes_dados.items():
    estado_maior, maior_precipitacao, estado_menor, menor_precipitacao = calcular_maior_menor_precipitacao(resultados)
    tabela_final.append([regiao, estado_maior, maior_precipitacao, estado_menor, menor_precipitacao])

#tabela final
tabela_final_df = pd.DataFrame(tabela_final, columns=["Região", "Estado Maior Precipitação Média por Estação", "Maior Precipitação Média (mm)", "Estado Menor Precipitação Média por Estação", "Menor Precipitação Média (mm)"])
print(tabela_final_df)
```
